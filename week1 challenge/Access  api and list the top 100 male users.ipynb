# access https://randomuser.me/ api and list the top 100 male users. 


```python
# Define the API endpoint and parameters
import requests
import json
```

```python
api_endpoint = "https://randomuser.me/api/"
params = {
    "results": 100,
    "gender": "male",
}

# Send a GET request to the API endpoint with the parameters
response = requests.get(api_endpoint, params=params)

# Check if the response was successful (HTTP status code 200)
if response.status_code == 200:
    # Parse the JSON data in the response
    data = json.loads(response.content)

    # Extract the list of users from the data
    users = data["results"]
    for i, user in enumerate(users):
        name = f"{user['name']['title']} {user['name']['first']} {user['name']['last']}"
        print(f"{i+1}. {name}")
else:
    print(f"Error: {response.status_code}")
```

# In this code, we set the results parameter to 100 to retrieve 100 users, 
# set the gender parameter to "male" to get only male users
# We then send a GET request to the API endpoint with the parameters using the requests.get() method, 
# and parse the JSON data in the response using the json.loads() method.
# Finally, we extract the list of users from the data and print the name of each user using a for loop. 
# The enumerate() function is used to get both the index and the value of each user in the loop.
